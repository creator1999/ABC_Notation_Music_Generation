# In this notebook I will visualize the model's performance.

##### Importing the required libraries and converting the csv file to pandas dataframe.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

model_csv = pd.read_csv('training_log.csv')
```

```python
model_csv.head()
```

##### Describing the loss and accuracy over the epochs.

```python
model_csv['loss'].describe()
```

```python
model_csv['acc'].describe()
```

##### Plotting the loss and accuracy of model as the function of epoch.

```python
sns.set()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(model_csv['epoch'], model_csv['loss'])
plt.title('Training Loss over Epochs')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(model_csv['epoch'], model_csv['acc'])
plt.title('Training Accuracy over Epochs')
plt.show()
```
